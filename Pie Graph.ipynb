```python
plt.pie(stds, labels = smedia, explode =[0,0,0,0.05,0])
plt.show()
```

```python
e = [0,0.05,0,0.05,0]
```

```python
e
```

```python
plt.pie(stds, labels = smedia, explode = e)
plt.show()
```

```python
plt.pie(stds, labels = smedia, explode = e, colors = ['b','y','r','green','m'])

plt.show()
```

```python
c = ['b','y','r','green','m']
plt.pie(stds, labels = smedia, explode = e, colors = c)

plt.show()
```

```python
plt.pie(stds, labels = smedia, explode = e,colors = c,  autopct= "%0.2F%%")
plt.show()
```

```python
plt.pie(stds, labels = smedia, explode = e,colors = c,  autopct= "%0.2F%%",radius=1.5)
plt.show()
```

```python
plt.pie(stds, labels = smedia, explode = e,  autopct= "%0.2F%%",radius=1.5,shadow = True)
plt.show()
```
