```python
days = np.arange(1,16,1)
```

```python
days
```

```python
temp = np.random.randint(20,35,(15))
```

```python
temp
```

```python
humid = np.random.randint(40,55,(15,))
```

```python
humid
```

```python
plt.bar(days,temp, color = "skyblue", linewidth =2, edgecolor = "red")
plt.show()
```

```python
plt.bar(days,humid, color = "yellow", linewidth = 2, edgecolor = "r")
plt.show()
```

```python
plt.bar(days,humid, color = "yellow", linewidth = 2, edgecolor = "r")
plt.bar(days,temp, color = "skyblue", linewidth =2, edgecolor = "red")
plt.show()
```

```python
plt.bar(days,humid, color = "yellow", linewidth = 2, edgecolor = "r", width = 0.90)
plt.bar(days,temp, color = "skyblue", linewidth =2, edgecolor = "red",width = 0.60)
plt.show()
```

```python
plt.bar(days,humid, color = "yellow", linewidth = 2, edgecolor = "r", width = 0.90)
plt.bar(days,temp, color = "skyblue", linewidth =2, edgecolor = "red",width = 0.60)
plt.legend(['Humidity','Temperature'],loc = 2)
plt.show()
```
